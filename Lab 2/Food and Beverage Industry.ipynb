```python
# Library imports for analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loding the dataset to be analyzed
dataset = pd.read_csv("C:/Users/Enzo/Downloads/Food Industry/Food_Data.csv")
```

```python
# Displaying the first 5 members of the dataset
dataset.head()
```

```python
# Displaying the last 5 members of the dataset
dataset.tail()
```

```python
# A description of the counts involved in the dataset
dataset.describe()
```

```python
# Structural information of the dataset
dataset.info()
```

```python
# Creating a line chart for visualization

plt.plot(dataset["Net_Sales"], dataset["Sales_Qty"])
plt.grid(True)
plt.xlabel("Net Sales")
plt.ylabel("Sales Quantity")
plt.title("Sales quantity vs Net sales")

# Show plot
plt.show()
```

```python
# Presenting this in form of a bar chart
plt.figure(figsize=(18, 6))
plt.bar(dataset["Net_Sales"], dataset["Sales_Qty"])
plt.grid(True)

# Add formating to the data
plt.xlabel("Net Sales")
plt.ylabel("Sales Quantity")
plt.title("Sales quantity vs Net sales")

# Show plot
plt.show()
```

```python
# Using seaborn, we can visualize the sale quantity of foods by category
sns.pairplot(dataset[['Price', 'Gross_sales', 'Net_Sales', 'Category']], hue='Category', palette='viridis')
plt.show()
```
